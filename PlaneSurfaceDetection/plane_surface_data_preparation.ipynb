```
import glob
surface_planes = '/content/drive/My Drive/computer_vision/capstone_project/EVA5_AI_Projects_new-master/data/datasets/capstone_dataset/ground_truths/planar_surface_gt_images/'
npy_files = glob.glob(surface_planes+"/*.npy")
```

```
for file_path in train_data:
  src_dir_name = '/content/drive/My Drive/computer_vision/capstone_project/EVA5_AI_Projects_new-master/data/datasets/capstone_dataset/ground_truths/planar_surface_gt_images/'
  tar_dir_name = '/content/drive/My Drive/computer_vision/EVA5_Capstone/YoloV3master/data/customdata/ground_truth/SurfacePlanes/'
  filename = os.path.splitext(file_path)[0].split("/")[-1]
  final_filename = filename+'_plane_parameters.npy'
  tar_abs_path = tar_dir_name+final_filename
  src_abs_path = src_dir_name+final_filename
  if os.path.exists(src_abs_path):
    print(f"Copying {src_abs_path}")
    shutil.copy(src_abs_path, tar_abs_path)


```
